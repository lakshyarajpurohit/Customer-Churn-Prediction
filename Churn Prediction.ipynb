### Imbalanced dataset is a dataset where target class is not distributed unifromly

```python
X = data.drop('Exited',axis = 1)
y = data['Exited']
```

### Oversampling data 

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res, y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size = 0.2, random_state = 42)
## If you don’t use stratify, the split might randomly allocate more samples from one class into either train or test, leading to biased results.
##  Ensures both train and test sets have the same class distribution as the original dataset.
```

### Feature Scaling (Useful in models where distance is required)

```python
from sklearn.preprocessing import StandardScaler    ## as we see credit score value and age
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test  = sc.transform (X_test)
```

```python
X_train
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred1)
```

```python
accuracy_score(y_test, y_pred1)
```

### here in imbalaced dataset we should not use accuracy as performance metrics and major chances are there for wrong performance as we see here 

## so we should handle imbalanced dataset by oversampling and uhdersampling

### oversampling - oversample minority class with replacement 
### undersampling - randomly delete rows from majority class to match with minority class but with this many imp info is lost 

### SMOTE - it generates randomly synthetic data based on linear interpolation so that data can be evenly distributed and not duplicate data , data with slight changes
